```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
import plotly.express as px
```

```python
# Q2
df = pd.read_csv("/content/drive/MyDrive/data set/wheetseeds.csv")
```

```python
# Q3
# take spacific column
x = df.loc[:,["Area","Perimeter","Compactness","Class"]]
y = x.to_numpy()
print(y)
```

```python
# Q4
# Separate data by class
class1 = x.loc[x.Class == 1,["Area","Perimeter","Compactness"]].to_numpy()
class2 = x.loc[x.Class == 2,["Area","Perimeter","Compactness"]].to_numpy()
class3 = x.loc[x.Class == 3,["Area","Perimeter","Compactness"]].to_numpy()

# Create a 3D plot
fig = plt.figure(figsize = (12,8))
ax = fig.add_subplot(projection='3d')

# Plot data points with different shapes
ax.scatter(*zip(*class1), marker='o', label='Class 1 (Circle)')
ax.scatter(*zip(*class2), marker='s', label='Class 2 (Square)')
ax.scatter(*zip(*class3), marker='^', label='Class 3 (Triangle)')

# Set labels for each axis
ax.set_xlabel('Area')
ax.set_ylabel('Perimeter')
ax.set_zlabel('Compactness')

ax.legend()
plt.show()
```

```python
# Q5
# Separate data by class
class1 = x.loc[x.Class == 1,["Area","Perimeter","Compactness"]].to_numpy()
class2 = x.loc[x.Class == 2,["Area","Perimeter","Compactness"]].to_numpy()
class3 = x.loc[x.Class == 3,["Area","Perimeter","Compactness"]].to_numpy()

class1_mean = np.mean(class1, axis=0)
class2_mean = np.mean(class2, axis=0)
class3_mean = np.mean(class3, axis=0)

print(class1_mean)
print(class2_mean)
print(class3_mean)


# # Create a 3D plot
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(projection='3d')

# Plot data points with different shapes
ax.scatter(class1[:, 0], class1[:, 1], class1[:, 2], marker='o', label='Class 1', color="r")
ax.scatter(class2[:, 0], class2[:, 1], class2[:, 2], marker='s', label='Class 1', color="b")
ax.scatter(class3[:, 0], class3[:, 1], class3[:, 2], marker='^', label='Class 1', color="g")

# mean
ax.scatter(class1_mean[0], class1_mean[1], class1_mean[2], s=100, marker='o', color='black')
ax.scatter(class2_mean[0], class2_mean[1], class2_mean[2], s=100, marker='s', color='black')
ax.scatter(class3_mean[0], class3_mean[1], class3_mean[2], s=100, marker='^', color='black')

# Set labels for each axis
ax.set_xlabel('Area')
ax.set_ylabel('Perimeter')
ax.set_zlabel('Compactness')

ax.legend()
plt.show()
```
